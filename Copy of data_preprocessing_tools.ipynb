# Data Preprocessing Tools

## Importing the libraries



```
import numpy as np  # allows us to work with arrays
import matplotlib.pyplot as plt   # allows plotting charts and graphs (pyplot module of matplotlib library)
import pandas as pd # allows importing dataset and creating matrix of features and dependednt variable vector
```

## Importing the dataset

```
dataset = pd.read_csv('Data.csv') # creating dataframe from given data
x = dataset.iloc[:,:-1].values # matrix of features
y = dataset.iloc[:,-1].values  # vector of dependent variables
```

```
print(x)
```

```
print(y)
```

## Taking care of missing data

```
from sklearn.impute import SimpleImputer # importing simpleimputer class from impute module of sklearn library
imputer = SimpleImputer(missing_values=np.nan, strategy = 'mean')   #object of the class
imputer.fit(x[:,1:3])   #applying fit method fit the strategy to x
x[:,1:3] = imputer.transform(x[:,1:3]) # updating x using strategy
```

```
print(x)
```

## Encoding categorical data

### Encoding the Independent Variable

```
from sklearn.compose import ColumnTransformer #import columntransformer class
from sklearn.preprocessing import OneHotEncoder #import encoder class

#creating object of class
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [0])], remainder='passthrough')  #passthrough ensures non encoded columns are preserved

#connecting object to matrix of features x
x = np.array(ct.fit_transform(x))
```

```
print(x)
```

### Encoding the Dependent Variable

```
from sklearn.preprocessing import LabelEncoder

#creating object of class
le = LabelEncoder()

y = le.fit_transform(y)
```

```
print(y)
```

## Splitting the dataset into the Training set and Test set

```
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test  = train_test_split(x,y, test_size = 0.2, random_state = 1)
```

```
print(x_train)
```

```
print(x_test)
```

```
print(y_train)
```

```
print(y_test)
```

## Feature Scaling

```
# importing standard scaler class
from sklearn.preprocessing import StandardScaler

# creating object of class

sc = StandardScaler()
 
# standardisation not applied to dummy variables since it would mean losing information given by the data
# eg. 1 0 0 is france but after standardisation it will be meaningless 

# only apply feature scaling on numerical values (eg. age, salary)

# fitting scaler on training set

x_train[:,3:] = sc.fit_transform(x_train[:,3:])

# transform test set features with same scaler (only transform not fit)

x_test[:,3:] = sc.transform(x_test[:,3:])



 
```

```
print(x_train)
```

```
print(x_test)
```
